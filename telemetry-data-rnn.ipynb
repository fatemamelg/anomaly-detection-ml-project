```python
busVolt = pd.read_csv('./Data/BusVoltage.csv')
busVolt.columns = ['Time', 'Bus Voltage']
series = busVolt['Bus Voltage']
#time = busVolt['Time']
time = pd.to_datetime(busVolt['Time'], format='%Y-%m-%d')

fig = plt.figure(figsize=(10, 5))
graph = fig.add_axes([0.1, 0.1, 0.8, 0.8])
graph.plot(time, busVolt['Bus Voltage'])
plt.show()
```
