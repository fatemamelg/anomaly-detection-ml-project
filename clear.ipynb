```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#from statsmodels.tsa.seasonal import seasonal_decompose
#import statsmodels.graphics.tsaplots as sgt
```

```python
# get the original data 
df = pd.read_csv('BatteryTempCPV6Averaged.csv')
# conver date (string ) to (date)
df['time'] = pd.to_datetime(df['time'])
# make the date as the index of data
df.set_index('time',inplace=True)
# make the period time is 5 min
df = df.asfreq('300s')
# fill the mising values with the mean
df = df[df != 0]
df = df.fillna(method='bfill')
df = df.fillna(value=df.mean())
df.isna().sum()
# get the avrage of an hour in column
df = df.resample('H').mean()
df.reset_index(inplace=True)
plt.plot(df['time'],df['TCPV6T (C)'])
plt.rcParams['figure.figsize'] = [20,5]
plt.ylim((-10,45))
plt.show()

```
